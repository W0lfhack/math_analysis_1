$$x_n = r x_n (1 - x_n)$$
$$x_n (1 - r(1 - x_n)) = 0 \implies \left[\begin{array}{l} x_n = 0 \\ 1 - r + r x_n = 0 \end{array} \right.  \implies \left[\begin{array}{l} x_n = 0 \\ x_n = 1 - \frac{1}{r} \end{array} \right.$$
Отображение имеет две неподвижные точки: $0$ и $1 - \frac{1}{r}$

$$1 - \frac{1}{r} \in [0;1] \implies \left\{\begin{array}{l}1 - \frac{1}{r} \ge 0 \\1 - \frac{1}{r} \le 1\end{array}\right.  \implies r \ge 1 $$

Найдем при каких $r$ неподвижные точки $0$ и $1 - \frac{1}{r}$ совпадают:

Значит отображение имеет одну неподвижную точку $0$ при $r \in [0;1]$ и две неподвижные точки $0$ и $1 - \frac{1}{r}$ при $r \in (1; 4]$.

Задача 2 \
Докажите, что при $x_0 \in (0; 1)$ и $r \in (0; 1]$ последовательность {$x_n$}, заданная логистическим отображением, монотонно убывает. Существует ли предел у данной последовательности при $r \in (0; 1]$? Докажите. Покажите графически.

Докажем, что последовательность строго убвает. Найдем отношение $\frac{x_{n+1}}{x_n}$:
$$\frac{x_{n+1}}{x_n} = \frac{r x_n (1 - x_n)}{x_n} = r (1 - x_n)$$
Раннее доказывалось, что если $0 < x_0 < 1$, то $0 < x_n < 1$ $\forall n \in \mathbb{N}, \, \forall r \in (0; 1)$. Так как $x_n \in (0;1)$, то и $1 - x_n \in (0;1)$
$$\left\{\begin{array}{l}r \in (0;1] \\ 1 - x_n \in (0;1)\end{array}\right. \implies r (1 - x_n) \in (0;1)$$
Значит $x_{n+1} < x_n$. Следовательно, последовательность строго убывает. \
$\forall n \in \mathbb{N}: x_n \in (0;1) \implies$ последовательность ограничена снизу нулем.

{$x_n$} - убывающая и ограниченная снизу последовательность. Значит, у такой последовательности есть предел. Найдем чему он равен. \
Пусть $\lim\limits_{n \to \infty} {x_n} =L$. Тогда $\lim\limits_{n \to \infty} {x_{n+1}} = \lim\limits_{n \to \infty} r x_n (1 - x_n) = L$. Получаем:
$$r L (1 - L) = L \implies L(r - r L - 1) = 0 \implies \left[\begin{array}{l} L = 0 \\ L = 1 - \frac{1}{r} \end{array} \right.$$
При $r \in (0;1): 1 - \frac{1}{r} < 0$, чего быть не может. При $r = 1: L = 0$. Значит, $\lim\limits_{n \to \infty} {x_n} =0$.


```python
import matplotlib.pyplot as plt


def f1(r, x_0):
    x = [x_0]
    n = [0]
    for i in range(1, 50):
        current = r * x[-1] * (1 - x[-1])
        x.append(current)
        n.append(i)
    plt.plot(n, x, label=f'r = {r}')

for r in range(2, 12, 2):
    r /= 10
    f1(r, 0.5)
plt.ylabel('x_n')
plt.xlabel('n')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

## Доказательство монотонности подпоследовательностей

### Дано

$r \in (2; 3)$, $f(x) = rx(1-x)$, $x^* = 1 - \dfrac{1}{r}$

$x_{2n} > x^*$, $x_{2n+1} < x^*$

---

### Ключевая формула

$$f(a) - f(b) = r(a - b)(1 - a - b)$$

---

### Лемма: $x_n > \frac{1}{2}$ для всех членов (после переходного процесса)

При $r \in (2; 3)$:

$$x^* = 1 - \frac{1}{r} > 1 - \frac{1}{2} = \frac{1}{2}$$

Так как $f(x^*) = x^*$ и $f$ непрерывна, в окрестности $x^*$ имеем $f(x) > \frac{1}{2}$. 

Последовательность сходится к $x^*$, значит все члены (после переходного процесса) больше $\frac{1}{2}$.

---

### Следствие: $(1 - a - b) < 0$

Для любых двух членов $a$ и $b$ одной чётности:

$$a > \frac{1}{2}, \quad b > \frac{1}{2} \quad \Rightarrow \quad a + b > 1 \quad \Rightarrow \quad (1 - a - b) < 0$$

---

### Доказательство

$$x_{n+2} - x_n = r(x_{n+1} - x_{n-1}) \cdot \underbrace{(1 - x_{n+1} - x_{n-1})}_{< 0}$$

Следовательно:

$$\mathrm{sign}(x_{n+2} - x_n) = -\mathrm{sign}(x_{n+1} - x_{n-1})$$

Разности чётных и нечётных членов имеют **противоположные знаки**. 

---

### Вывод

- $x_{2n} > x^*$ и сходится к $x^*$ сверху $\Rightarrow$ **убывает**
- Значит $x_{2n+1}$ $\Rightarrow$ **возрастает**

$$\boxed{x_{2n} \downarrow x^*, \quad x_{2n+1} \uparrow x^*}$$

Задача 3 \
Пусть $r \in (2; 3)$, $x_{2n} > x^*$, $x_{2n+1} < x^*$. Что вы можете сказать о монотонности подпоследовательностейa {$x_{2n}$}, {$x_{2n+1}$}? Докажите. Проверьте графически.

Рассмотрим последовательность {$x_{2n}$}, $x_{2n} > x^*$. Найдем чему равно отношение $\frac{x_{2n+2}}{x_{2n}}$:
$$\frac{x_{2n+2}}{x_{2n}} = \frac{r x_{2n+1} (1 - x_{2n+1})}{x_{2n}} = \frac{r^2 x_{2n} (1 - x_{2n}) (1 - r x_{2n} (1 - x_{2n}))}{x_{2n}} = r^2 (1 - x_{2n}) (1 - r x_{2n} (1 - x_{2n})) = r^2 (1 - r x_{2n} + 2rx_{2n}^2 - x_{2n} - rx_{2n}^3) = r^2 f(x_{2n})$$
Продеффиринцируем $f(x_{2n})$:
$$f'(x_{2n}) = -3rx_{2n}^2 + 4rx_{2n} - (r + 1) $$
$$D = 16r^2 - 12r(r + 1) = 4r(r - 3)$$
$$\left\{\begin{array}{l}r \in (2; 3) \\D = 4r(r -3)\end{array}\right. \implies D < 0 \implies f'(x_{2n}) < 0$$
Функция $f(x_{2n})$ монотонно убывает. $x_{2n} > x^* = 1 - \frac{1}{r} \implies x_{2n} \in (1 - \frac{1}{r}; 1)$.
$$\left\{\begin{array}{l}x_{2n} \in (1 - \frac{1}{r}; 1)\\ \frac{x_{2n+2}}{x_{2n}} = r^2 f(x_{2n})\\ r^2 f(1 - \frac{1}{r}) = r^2 \frac{1}{r^2} = 1 \end{array}\right. \implies \frac{x_{2n+2}}{x_{2n}} < 1, \, \forall x_{2n} \in (1 - \frac{1}{r}; 1)$$
Значит, последовательность {$x_{2n}$} - монотонно убывает. \
Аналогично рассмотрими последовательность {$x_{2n+1}$}:
$$\left\{\begin{array}{l}x_{2n+1} \in (0;1 - \frac{1}{r})\\ \frac{x_{2n+3}}{x_{2n+1}} = r^2 f(x_{2n+1})\\ r^2 f(1 - \frac{1}{r}) = r^2 \frac{1}{r^2} = 1 \end{array}\right. \implies \frac{x_{2n+3}}{x_{2n+1}} > 1, \, \forall x_{2n+1} \in (0;1 - \frac{1}{r})$$
Значит, последовательность {$x_{2n+1}$} - монотонно возрастает.

```python
import matplotlib.pyplot as plt


def f1(r, x_0):
    x = [x_0]
    n = [0]
    x_2n = [x_0]
    i_2n = [0]
    x_2n_1 = []
    i_2n_1 = []
    const = [1 - 1/r]
    for i in range(1, 10):
        const.append(1 - 1/r)
        current = r * x[-1] * (1 - x[-1])
        x.append(current)
        n.append(i)
        if i % 2 == 0:
            x_2n.append(current)
            i_2n.append(i)
        else:
            x_2n_1.append(current)
            i_2n_1.append(i)
    plt.title(f'r = {r}, x_0 = {x_0}')
    plt.plot(n, const, label=str('x*'))
    plt.plot(n, x, label=str('x_n'))
    plt.plot(i_2n, x_2n, label=str('x_2n'))
    plt.plot(i_2n_1, x_2n_1, label=str('x_2n_1'))
    plt.ylabel('x_n')
    plt.xlabel('n')
    plt.legend(bbox_to_anchor=(1, 1))
    plt.show()

f1(2.2, 0.6)
f1(2.5, 0.7)
f1(2.7, 0.7)
f1(2.9, 0.7)
```
