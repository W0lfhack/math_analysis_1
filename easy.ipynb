Дано следующее одномерное логистическое отображение:
$$x_{n+1} = r x_n (1 - x_n)$$

Задача 1

Докажите, что $\forall n \in \mathbb{N}, \, \forall r \in (0; 1]$  
$$0 < x_0 < 1 \implies 0 < x_n < 1$$

Доказательство:

Докажем с помомощью метода математической индукции. Так как $0 < x_0 < 1$, то $0 < 1 - x_0 < 1$. Также известно, что $0 < r \leq 1$. Значит $x_1 = r x_0 (1 - x_0) \in (0; 1)$, потому что все множители $r, x_0$ и $1 - x_0 \in (0; 1]$. Положим это за базис индукции.


Докажем в общем виде, что если $0 < x_{n-1} < 1$, то $0 < x_n < 1$. Доказательство аналогичное: $r, x_{n-1}$ и $1 - x_{n-1} \in (0; 1] \implies 0 < x_n < 1$, что и требовалось доказать.

Задача 2

Сделайте вывод: как параметр $r$ влияет на поведение функции зависимости $x_n$ от $x_{n-1}$? Постройте эту функцию для нескольких различных значений $r$.

```python
import matplotlib.pyplot as plt


def f1(r):
    x_n = []
    x_last = []
    y = []

    for x in [i/100 for i in range(1, 100)]:
        y.append(x)
        current = r * x * (1 - x)
        x_n.append(current)
        x_last.append(x)

    plt.title(f'r = {r}')
    plt.ylabel('$x_n$')
    plt.xlabel('$x_{n-1}$')
    plt.plot(x_last, x_n)
    plt.plot(x_last, y)
    plt.show()


f1(0.5)
f1(1)
f1(1.5)
f1(2)
f1(2.5)
f1(3)
f1(3.5)
f1(3.5699456)
f1(4)
```

При малых $r \in (0; 1]$ отображение будет монотонно сходится к $0$, так как $0$ - это единственная неподвижная точка. \
При $r \in (1; 4)$ отображение имеет две неподвижные точки. Также с ростом r отображение становится более хаотичным.

Дана следующая функция:
$$g(x_{n+1}) = r x_n (1 - x_n^2), \, r \in [0; \frac{3\sqrt3}{2}]$$

Задача 3

Для заданной вариантом функции $g(x_n)$:
1. Постройте графики зависимости $x_n$ от $x_{n-1}$ для нескольких различных значений $r$.
2. Сделайте вывод о сходстве или различии поведения логистического отображения и точечного отображения из вашего варианта. Предположите: чем могут быть вызваны сходства/различия?

```python
import matplotlib.pyplot as plt


def f2(r):
    x_n = []
    x_last = []
    y = []

    for x in [i/100 for i in range(1, 100)]:
        y.append(x)
        current = r * x * (1 - x**2)
        x_n.append(current)
        x_last.append(x)

    plt.title(f'r = {r}')
    plt.ylabel('$x_n$')
    plt.xlabel('$x_{n-1}$')
    plt.plot(x_last, x_n)
    plt.plot(x_last, y)
    plt.show()


f2(0)
f2(0.5)
f2(1)
f2(1.5)
f2(2)
f2(2.5)
f2(3**1.5 / 2)
```

Сходства: \
При $r \in (0; 1)$ логистическое отображение и функция имеют одну неподвижную точку $0$ и монотонно к ней сходятся. \
Различия: \
Разное максимальное значение $r$, при которых отображения остаются замкнутыми. Для логистического отображения - это $r = 4$, для функции $g(x_n)$ это - $r = \frac{3\sqrt{3}}{2}$. \
Сходства и различия возникают из-за разной степени полиномов.
