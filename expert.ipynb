###Задача 1
Следует ли асимптотическая устойчивость $ x^* $ из условия
$$ \exists \delta_0 > 0: |x_0 - x^*| < \delta_0 \Longrightarrow x_n \xrightarrow{n \to \infty} x^* ?$$
Обоснуйте свой ответ.

### Доказательство:
Рассмотрим отображение при $x_0 \in (-1; 1)$:

$$f(x_n) =
\begin{cases}
0, \ x_{n-1} = 0  \\
\frac{1}{|x_{n-1}|}, \ |x_{n-1}| ≠ 0, \ n = 1\\
\frac{|x_{n-1}|}{2}, \ n \geq 2
\end{cases}$$

Так как $0 < |x_0| < 1$, то $x_1 = \frac{1}{|x_{n-1}|} > 1$. А далее все последующие члены будут вдвое меньше предыдущего и будут иметь вид: $x_n = \frac{x_1}{2^{n-1}}$. Так как $n \to \infty$, то $x_n \to 0$. Значит $ f(x_n)$ обладает свойством сходимости к усточивой точке $ x^* = 0 $. Для асимптотической устойчивости должен существовать «коридор» вокруг $ x^* $ и  $ x_n $ находится в этом коридоре при любых $n$, то есть:

$$ \forall \varepsilon > 0 \ \exists \delta > 0: |x_0 - x^*| < \delta \Longrightarrow |x_n - x^*| < \varepsilon \ \forall n $$

Но для $ \varepsilon = 1 $ и любого $ \delta > 0 $ можно выбрать $ 0 < x_0 < min(1,\delta) $, тогда $ x_1 = \frac{1}{x_0} > 1 = \varepsilon$. Условие сходимости выполняется, но $ x_1 > \varepsilon $, что не удовлятворяет условию устойчивости.

## Задача 2
Докажите или опровергните утверждение.

При $ r \in (0; 1)$ неподвижная точка $x^* = 0$ является устойчивой. Является ли она асимптотически устойчивой?

### Доказательство:

Из первой задачи уровня Easy, доказано, что $\forall n \in \mathbb{N}, \, \forall r \in (0; 1]$  
$$0 < x_0 < 1 \implies 0 < x_n < 1$$

Из второй задачи уровня Easy доказана монотонная сходимость к 0 при $r \in (0, 1)$. По теореме Вейерштрасса у монотонно убывающей ограниченной последовательности существует предел $a \in \mathbb{R}$. Тогда
$$ a = ra(1 - a) \implies a(1 - r(1 - a)) = 0 $$
$ a = 0 $ или $ a = 1 - \frac{1}{r} $. Но $ r \in (0; 1) \implies \frac{1}{r} > 1 \implies 1 - \frac{1}{r} < 0$, что невозможно при $x_n \geq 0$. Значит, предел $ a = 0 $, то есть $x_n \rightarrow 0$ при любом $x_0 \in [0; 1]$.

**Устойчивость по Ляпунову:**

Возьмём произвольное $\varepsilon > 0$ и положим $\delta = \min\{\varepsilon, 1\}$. Если $|x_0 - 0| < \delta$, то $0 \leq x_0 < \delta \leq 1$.

Поскольку $x_{n+1} = rx_n(1-x_n) < rx_n < x_n$ (последовательность убывает), имеем:
$$|x_n - 0| = x_n \leq x_0 < \delta \leq \varepsilon \quad \text{для всех } n.$$


**Асимптотическая устойчивость:**

По определению асимптотической устойчивости: $$ \exists \delta_0 > 0: |x_0 - x^*| < \delta_0 \Longrightarrow x_n \xrightarrow{n \to \infty} x^* $$

Положим $\delta_0 = 1$. При $|x_0| < 1$ имеем $x_n \to 0$ (доказано выше).

Таким образом, неподвижная точка $x^* = 0$ является **асимптотически устойчивой**.

## Задача 3
Докажите, что точка $x^* = 0$ при $r \in (2; 3)$ является неусточивой.

### Доказательство:
**Неустойчивость**
$$\exists \varepsilon > 0 \ \forall \delta > 0, \ |x_0 - x^*| < \delta \implies |x_n - x^*| \geq \varepsilon$$

Выделим полный квадрат в логистическом отображении:
$$x_1 = rx_0(1-x_0) = -r(x_0 - \frac{1}{2})^2 + \frac{r}{4} \leq \frac{r}{4}$$


Рассмотрим два случая $\delta \leq \frac{1}{2}$ и $\delta > \frac{1}{2}$:
1) $\delta \leq \frac{1}{2}$: \
В easy мы строили логистическое отображеие для различных значений $r$. И можно заметить, что при $r \in (2;3), \ x_0 \in (0;\frac{1}{2})$ значение $x_1 = rx_0(1-x_0)$ будет всегда больше, чем $x_0$. Если положить $\varepsilon = x_0$, то $x_1 > x_0 = \varepsilon, \ \forall \delta \leq \frac{1}{2}$
2) $\delta > \frac{1}{2}$: \
Положим $\varepsilon = \frac{r}{4}$. Так как $\delta > \frac{1}{2}$, то мы всегда можем выбрать $x_0 = \frac{1}{2}$, а значение $x_1 = \frac{r}{4} \geq \varepsilon$


Тогда $\forall \delta > 0$ найдётся $\varepsilon$ и $x_0$ для которых $|x_0 - x^*| < \delta \implies |x_1 - x^*| \geq \varepsilon$, что удовлетворяет условию неустойчивости.

### Задача 4
Напишите функцию, которая для заданных $x_0, y_0 = x_0 + \varepsilon, r$ строит две траектории лестницы Ламерея на одном графике. Постройте этот график для $r = 4$. Как можно интерпретировать результаты?

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, r) -> float:
    """
    Базовое логистическое отображение

    Параметры:
    x : Значение переменной x
    r : Значение переменной r
    """
    return r * x * (1 - x)
```

```python
r = 4
e = 0.01
x_0 = 0.2
y_0 = x_0 + e
x0_coords = np.linspace(0, 1, 500)
y0_coords = [f(x0_coords[i], r) for i in range(500)]

x1 = x_0
x_coords_x0 = [x_0]
y_coords_x0 = [0]

for _ in range(10):
        xn_x0 = f(x1, r)

        x_coords_x0.append(x1)
        y_coords_x0.append(xn_x0)

        x_coords_x0.append(xn_x0)
        y_coords_x0.append(xn_x0)

        x1 = xn_x0

x2 = y_0
x_coords_y0 = [y_0]
y_coords_y0 = [0]

for _ in range(10):
        xn_y0 = f(x2, r)

        x_coords_y0.append(x2)
        y_coords_y0.append(xn_y0)

        x_coords_y0.append(xn_y0)
        y_coords_y0.append(xn_y0)

        x2 = xn_y0

plt.figure(figsize=(10, 10))
plt.plot(x0_coords, y0_coords, label='f(x)')
plt.plot([0, 1], [0, 1], 'k--', label='y = x')
plt.plot(x_coords_x0, y_coords_x0, 'red', label=f'x_0 = {x_0}')
plt.plot(x_coords_y0, y_coords_y0, 'orange', label=f'y_0 = {y_0}')
plt.title(f'r = {r}')
plt.xlabel('x_n')
plt.ylabel('x_n+1')
plt.legend()
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

Можно заметить, что при небольшом измении начального условия оранжевая тракетория заметно отклонилась от красной. Это называется эффектом бабочки.

### Задача 5
1. Постройте бифуркационную диаграмму логистического отображения.
2. Проанализируйте: как интерпретировать полученный
график?
3. Где на диаграмме находится $r_{\infty}$? Как ведет себя система
до $r_{\infty}$? После?

```python
r_values = np.linspace(0, 4, 1000)

R = []
X = []
for x0 in [0.1, 0.2, 0.4, 0.5, 0.6, 0.9]:
    for r in r_values:
        x = x0
        for _ in range(1400):
            x = f(x, r)

        for _ in range(100):
            x = f(x, r)
            R.append(r)
            X.append(x)

plt.figure(figsize=(12, 5))
plt.plot(R, X, ',', alpha=0.5)
plt.axvline(3.5699456, color='red')
plt.xlabel('r')
plt.ylabel('x')
plt.title('Бифуркационная диаграмма')
plt.show()
```

При $r \in (0; 1] \ x_n$ стремится к неподвижной точке $0$ \
При $r \in (1; 3] \ x_n$ стремится к неподвижной точке $1 - \frac{1}{r}$ \
При $r \in (3; r_{\infty})$ наблюдаются циклы длиной 2, 4, 8 ... \
При $r \in [r_{\infty}; 4)$ система не иммет усточивого цикла. Есть окна периодичности.

### Задача 9
Для заданного вариантом отображения $g(x_n)$:
1. Численно или аналитически найдите верхнюю и нижнюю границы параметра $r$, при котором точка $x^* = 0 $ является устойчивой, неустойчивой;
2. Постройте бифуркационную диаграмму. Отметьте сходства или различия с диаграммой логистического отображения;
3. Визуализируйте окна периодичности, если они есть.

$$g(x_{n+1}) = rx_n(1-x_n^2), \ r \in [0; \frac{3\sqrt3}{2}]$$

```python
def g(x, r):
    """
    отображение, заданное вариантом

    Параметры:
    x : Значение переменной x
    r : Значение переменной r
    """
    return r * x * (1 - x**2)
```

1. Чтобы точка $x = 0$ была устойчивой последовательность должна монотонно убывать и сходиться к ней. Значит, $\frac{x_{n+1}}{x_n} < 1$:
$$\frac{x_{n+1}}{x_n} = \frac{r x_n (1 - x_n^2)}{x_n} = r (1 - x_n^2) < 1 \implies r < \frac{1}{1 - x_n^2}$$
$$\left\{\begin{array}{l}x_n \to 0 \\ r < \frac{1}{1 - x_n^2}\end{array}\right. \implies r < 1 \implies r \in [0; 1)$$
При $r \in [0; 1)$ точка $x^* = 0$ является устойчивой, при $r \in (1; \frac{3\sqrt3}{2}]$ - неустойчивой.

2. Построим бифуркационную диаграмму для отображения $g(x)$:

```python
r_values = np.linspace(0, 3**1.5 / 2, 1000)

R = []
X = []
for x0 in [0.1, 0.2, 0.4, 0.5, 0.6, 0.9]:
    for r in r_values:
        x = x0
        for _ in range(1400):
            x = g(x, r)

        for _ in range(100):
            x = g(x, r)
            R.append(r)
            X.append(x)

plt.figure(figsize=(12, 5))
plt.plot(R, X, ',', alpha=0.5)
plt.xlabel('r')
plt.ylabel('x')
plt.title('Бифуркационная диаграмма')
plt.show()
```

При $r \leq 1 $ наблюдается устойчивость точки $x^* = 0$ \
При $r > 1$ точка $x^* = 0 $ теряет устойчивость
